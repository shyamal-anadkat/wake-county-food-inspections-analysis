```python
import os
from dotenv import load_dotenv

import numpy as np
import pandas as pd
import csv
from pathlib import Path
import requests
```

```python
load_dotenv()
```

```python
#url https://crime-data-explorer.fr.cloud.gov/pages/docApi
weather_token = os.getenv('WEATHER_TOKEN_KEY')
type(weather_token)
```

```python
def get_wakecounty_stations():
    
    token = weather_token
   
    endpoint = 'https://www.ncdc.noaa.gov/cdo-web/api/v2/stations?'
    
    payload = {
        
        'datasetid':'GHCND',
        'locationid': 'FIPS:37183',
        'limit': 1000
    
    }
    
    headers = {'Token': token}
    
    response = requests.get(url = endpoint, params=payload ,headers = headers)
    
    response = response.json()

    return response['results']
stations = get_wakecounty_stations()
print(stations)
```

```python
WakeCounty_df = pd.DataFrame(stations)
coverage_df = WakeCounty_df[WakeCounty_df['datacoverage'] >= 1.0]
coverage_df

```

```python
def get_weather_2019():
    
    token = weather_token
    
    endpoint = 'https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GHCND'
    
    payload = {
        
        'datatypeid': 'TAVG',
        'locationid': 'FIPS:37183',
        'startdate': '2019-01-01',
        'enddate': '2019-12-31',
        'limit':'1000',
        'stationid': 'GHCND:USW00013722',
        'units':'standard'
    }
    
    headers = {'Token': token}
    
    r = requests.get(url = endpoint, params=payload ,headers = headers)
    
    r = r.json()

    return r['results']

def get_weather_2020():
    
    token = weather_token
    
    endpoint = 'https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GHCND'
    
    payload = {
        
        'datatypeid': 'TAVG',
        'locationid': 'FIPS:37183',
        'startdate': '2020-01-01',
        'enddate': '2020-12-31',
        'limit':'1000',
        'stationid': 'GHCND:USW00013722',
        'units':'standard'
    }
    
    headers = {'Token': token}
    
    r = requests.get(url = endpoint, params=payload ,headers = headers)
    
    r = r.json()

    return r['results']

def get_weather_2021():
    
    token = weather_token
    
    endpoint = 'https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GHCND'
    
    payload = {
        
        'datatypeid': 'TAVG',
        'locationid': 'FIPS:37183',
        'startdate': '2021-01-01',
        'enddate': '2021-10-21',
        'limit':'1000',
        'stationid': 'GHCND:USW00013722',
        'units':'standard'
    }
    
    headers = {'Token': token}
    
    r = requests.get(url = endpoint, params=payload ,headers = headers)
    
    r = r.json()

    return r['results']
    
```

```python
weather2019 = get_weather_2019()
print(weather2019)
```

```python
weather2020 = get_weather_2020()
print(weather2020)
```

```python
weather2021 = get_weather_2021()
print(weather2021)
```

```python
weather2019_df = pd.DataFrame(weather2019)
weather2020_df = pd.DataFrame(weather2020)
weather2021_df = pd.DataFrame(weather2021)
weather_df = pd.concat([weather2019_df,weather2020_df,weather2021_df])
weather_df['date'] = pd.to_datetime(weather_df['date'])
weather_df = weather_df.drop(['station','attributes'],axis=1)
weather_df = weather_df.pivot(index='date',columns='datatype', values='value')
weather_df
```
